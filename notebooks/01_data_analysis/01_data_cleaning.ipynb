# 🧹 01. Data Cleaning and Initialization

**Objective:** Load raw e-commerce sales datasets, handle missing values, duplicates, parse datatypes, detect outliers, and save cleaned tables.

## 1. Import Libraries
We start by importing the necessary python data science packages.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

plt.style.use('seaborn-v0_8-whitegrid')
print("Libraries imported successfully.")
```

## 2. Load Raw Datasets
Load raw datasets from the `data/raw` directory.

```python
data_dir = "../../data/raw"
customers = pd.read_csv(os.path.join(data_dir, "customers.csv"))
orders = pd.read_csv(os.path.join(data_dir, "orders.csv"))
products = pd.read_csv(os.path.join(data_dir, "products.csv"))
order_items = pd.read_csv(os.path.join(data_dir, "order_items.csv"))
ratings = pd.read_csv(os.path.join(data_dir, "ratings.csv"))
payments = pd.read_csv(os.path.join(data_dir, "payments.csv"))
returns = pd.read_csv(os.path.join(data_dir, "returns.csv"))
shipping = pd.read_csv(os.path.join(data_dir, "shipping.csv"))
inventory = pd.read_csv(os.path.join(data_dir, "inventory.csv"))
marketing = pd.read_csv(os.path.join(data_dir, "marketing.csv"))

print(f"Loaded {len(customers)} customers, {len(orders)} orders, {len(products)} products.")
```

## 3. Data Integrity & Missing Values
Inspect columns, data types, and check for missing/null values across key tables.

```python
print("Orders missing values:")
print(orders.isnull().sum())

# Example logic for handling missing values
# orders['order_delivered_date'] = pd.to_datetime(orders['order_delivered_date'])
# orders['order_delivered_date'].fillna(orders['order_estimated_delivery_date'], inplace=True)
```

## 4. Handling Duplicates
Identify and drop duplicated rows.

```python
print(f"Customer duplicates: {customers.duplicated().sum()}")
customers = customers.drop_duplicates()
```

## 5. Outlier Analysis
Identify statistical outliers in numerical fields such as price and freight value.

```python
# IQR Method example on products or items prices
q1 = order_items['price'].quantile(0.25)
q3 = order_items['price'].quantile(0.75)
iqr = q3 - q1
outliers = order_items[(order_items['price'] < (q1 - 1.5 * iqr)) | (order_items['price'] > (q3 + 1.5 * iqr))]
print(f"Number of price outliers: {len(outliers)}")
```

## 6. Save Cleaned Data
Export the cleaned datasets into the `data/interim` folder.

```python
interim_dir = "../../data/interim"
os.makedirs(interim_dir, exist_ok=True)
customers.to_csv(os.path.join(interim_dir, "cleaned_customers.csv"), index=False)
orders.to_csv(os.path.join(interim_dir, "cleaned_orders.csv"), index=False)
products.to_csv(os.path.join(interim_dir, "cleaned_products.csv"), index=False)
print("Cleaned datasets saved successfully.")
```
